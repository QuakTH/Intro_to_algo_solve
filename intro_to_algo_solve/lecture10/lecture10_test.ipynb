## Import modules

## Create a Hash Table instance

```python
ht = hash_table.HashTable()
```

## Insert multiple values

```python
for alphabet in 'ABCDEFGHIJKLMNOP':
    ht.insert(alphabet, random.randint(1, 100))

    print(f'-- After inserting {alphabet} --')
    ht.print_hash_table()
    print()
```

```python
ht.print_hash_table()
```

## Searching

```python
ht.search('A')
```

```python
ht.search('K').value
```

```python
ht.search('Z')
```

## Inserting duplicated key value

```python
ht.insert('D', -100)
ht.print_hash_table()
```

```python
keys = list('ABCDEFGHIJKLMNOPQY')
random.shuffle(keys)


for alphabet in keys:
    ht.delete(alphabet)

    print(f'-- After Deleting {alphabet} --')
    ht.print_hash_table()
    print()
```
